```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', np.nan, 'Female'],
    'Math_Score': [88, 92, 95, np.nan, 70, 85, 999, 77],
    'Science_Score': [90, 85, np.nan, 80, 65, 75, 60, 82],
    'GPA': [3.8, 3.6, 3.9, 3.4, 2.8, 3.2, 4.5, 3.1]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Math_Score'].fillna(df['Math_Score'].mean(), inplace=True)
df['Science_Score'].fillna(df['Science_Score'].mean(), inplace=True)
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
Q1 = df['Math_Score'].quantile(0.25)
Q3 = df['Math_Score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df[(df['Math_Score'] < lower) | (df['Math_Score'] > upper)]
```

```python
df.loc[df['Math_Score'] > upper, 'Math_Score'] = upper
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Apply MinMaxScaler only on GPA column and create a new column
df['GPA_scaled'] = scaler.fit_transform(df[['GPA']])
```
